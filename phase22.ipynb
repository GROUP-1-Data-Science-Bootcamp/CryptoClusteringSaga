## Importing Libraries

The provided code imports necessary libraries.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import plotly.graph_objects as go
```

The code loads a CSV file named "coins_data.csv" into a DataFrame object named df using pd.read_csv().
It then displays the first few rows of the DataFrame using df.head() to provide a preview of the data.


```python
df = pd.read_csv(r"C:\Users\Admin\Desktop\quera\v1\New folder\coins_data.csv")
df.head()
```

## Scatter Plot of Volume and MarketCap

The provided code snippet generates a scatter plot to visualize the relationship between the 'Volume' and 'MarketCap' variables in the dataset.

```python
data = df[['Volume', 'MarketCap']].copy()
plt.scatter(data['Volume'], data['MarketCap'])
plt.show()
```

## Data Scaling
This code standardizes the 'Volume' and 'MarketCap' variables using StandardScaler() and creates a scatter plot of the standardized data.


```python
standarder = StandardScaler()
standardized_data = standarder.fit_transform(data)
data = pd.DataFrame(standardized_data, columns=['Volume', 'MarketCap'])

plt.scatter(data['Volume'], data['MarketCap'])
plt.show()
```

## Making Dataset
It creates a new DataFrame named data by selecting the 'Volume', 'MarketCap', and 'ProofType' columns from the existing DataFrame df

```python
data = df[['Volume', 'MarketCap', 'ProofType']].copy()
data.head()
```

## Making Dendograms
The code generates dendrograms for various combinations of distance metrics and linkage methods, using the linkage() and dendrogram() functions.
The dendrograms visualize hierarchical clustering structures within the data DataFrame, providing insights into the relationships between data points

```python
metrics = ['braycurtis', 'canberra', 'chebyshev', 'cityblock', 'correlation', 'cosine', 'hamming']
methods = ['single', 'ward', 'complete', 'average', 'centroid', 'median']

for metric in metrics:
    for method in methods:
        try:
            linkage_data = linkage(data, method=method, metric=metric)
            dendrogram(linkage_data, orientation = 'top', distance_sort='descending', show_leaf_counts=True)

            plt.title(f'{metric}, {method}')
            plt.show()
        except:
            print(f"couldn't do {metric} and {method}")
```

the chebyshev metric provides better results 

## Converting Data
The code uses a LabelEncoder to convert the categorical values in the 'ProofType' column of the DataFrame df into numerical labels.

```python
label_encoder = LabelEncoder()
data['ProofType'] = label_encoder.fit_transform(df['ProofType'])
data
```

## Creating a 3D-plot
This code creates a 3d-scatter plot of the data

```python
fig = go.Figure(data=go.Scatter3d(
    x=data['Volume'],
    y=data['MarketCap'],
    z=data['ProofType'],
    mode = 'markers',
    marker = dict(
        sizemode = 'diameter',
        sizeref = 750,
        color='blue'
        )
))


fig.show()
```
