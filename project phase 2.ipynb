```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r"C:\Users\Admin\Desktop\quera\v1\New folder\Q1_data.csv")
df
```

```python
X = df[["market_cap" , "volume"]]
fig , axes = plt.subplots(1 , 1)
axes = plt.scatter(df["market_cap"] , df["volume"])
```

```python
clf = KMeans(n_clusters=5, random_state=0, n_init="auto")
clf.fit(X)
y_pred = clf.predict(X)
fig , axes = plt.subplots(1 , 1)
axes = plt.scatter(df["market_cap"] , df["volume"] , c= y_pred)
axes = plt.scatter(clf.cluster_centers_[:,0] , clf.cluster_centers_[:,1])
plt.colorbar(ticks=range(10))
plt.show()
```

```python
from sklearn.metrics import silhouette_score
k = np.arange(1,11)
distorsions = []
s = []
for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)
    
for i in range(1,len(k)):
    clf = KMeans(n_clusters=k[i])
    clf.fit(X)
    s.append(silhouette_score(X, clf.labels_, metric='euclidean'))
    print(f"k = {k[i]}, silhouette score is {silhouette_score(X, clf.labels_, metric='euclidean')})")
    

fig , axes = plt.subplots()
axes = plt.plot(k , distorsions)
axes = plt.plot(k[1:11],s)
```

```python
clf = KMeans(n_clusters=4, random_state=0, n_init="auto")
clf.fit(X)
y_pred = clf.predict(X)
fig , axes = plt.subplots(1 , 1)
axes = plt.scatter(df["market_cap"] , df["volume"] , c= y_pred)
axes = plt.scatter(clf.cluster_centers_[:,0] , clf.cluster_centers_[:,1])
plt.colorbar(ticks=range(10))
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
def plot_dendrogram(model, **kwargs):
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    dendrogram(linkage_matrix, **kwargs)
clf = AgglomerativeClustering(distance_threshold=0, n_clusters=None)  
clf.fit(X_1)  
plot_dendrogram(clf, truncate_mode="level")
```
